```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/imersao_python.xlsx", sheet_name = "Principal")
df_principal
```

```python
df_total_de_acoes = pd.read_excel("/content/imersao_python.xlsx", sheet_name = "Total de Ações")
df_total_de_acoes
```

```python
df_ticker = pd.read_excel("/content/imersao_python.xlsx", sheet_name = "Ticker")
df_ticker
```

```python
df_chatGpt = pd.read_excel("/content/imersao_python.xlsx", sheet_name = "ChatGpt")
df_chatGpt
```

```python
df_principal = df_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)": "Valor_final", "Var. Dia (%)": "Var_dia_pct"}).copy()
df_principal
```

```python
df_principal["Var_pct"] = df_principal["Var_dia_pct"]/100
df_principal["Valor_inicial"] = df_principal["Valor_final"]/(df_principal["Var_pct"] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_de_acoes, left_on="Ativo", right_on="Código", how="left")
df_principal
```

```python
df_principal = df_principal.drop(columns=["Código"])
df_principal
```

```python
df_principal["Variacao_reais"] = (df_principal["Valor_final"] - df_principal["Valor_inicial"]) * df_principal["Qtde. Teórica"]
df_principal
```

```python
pd.options.display.float_format = "{:.2f}".format
df_principal
```

```python
df_principal["Qtde. Teórica"] = df_principal["Qtde. Teórica"].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica":"Qtd_teorica"}).copy()
df_principal
```

```python
df_principal["Resultado"] = df_principal["Variacao_reais"].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how="left")
df_principal = df_principal.drop(columns=["Ticker"])
df_principal
```

```python
df_principal = df_principal.merge(df_chatGpt, left_on="Nome", right_on="Nome da Empresa", how="left")
df_principal = df_principal.drop(columns=["Nome da Empresa"])
df_principal
```

```python
df_principal = df_principal.drop(columns=["Idade em Anos_y"])
df_principal = df_principal.drop(columns=["Segmento_y"])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Segmento_x	":"Segmento", "Idade em Anos_x":"Idade" }).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Segmento_x":"Segmento"}).copy()
df_principal
```

```python
df_principal['Idade'] = pd.to_numeric(df_principal['Idade'], errors='coerce')
df_principal['Idade'].fillna(-1, inplace=True)
df_principal["Cat_idade"] = df_principal["Idade"].apply(lambda x: "Mais de 100" if x > 100 else ("Menos de 50" if x < 50 and x != -1 else "Entre 50 e 100"))
df_principal
```

```python
df_principal['Idade'] = df_principal['Idade'].astype(int)
df_principal
```

```python
maior = df_principal["Variacao_reais"].max()
menor = df_principal["Variacao_reais"].min()
media = df_principal["Variacao_reais"].mean()
media_subiu = df_principal[df_principal["Resultado"] == "Subiu"]["Variacao_reais"].mean()
media_desceu = df_principal[df_principal["Resultado"] == "Desceu"]["Variacao_reais"].mean()

# Imprimindo os resultados
print(f"Maior\tR$, {maior:,.2f}")
print(f"Menor\tR$, {menor:,.2f}")
print(f"Média\tR$, {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal["Resultado"] == "Subiu"]
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")["Variacao_reais"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("Resultado")["Variacao_reais"].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x="Resultado", y="Variacao_reais", text="Variacao_reais", title="Variacao Reais por Resultado")
fig.show()
```
